```python
# thereâ€™s no separate function called bubbleplot in Matplotlib.

# You just create a bubble plot using plt.scatter() by setting: the size of the markers (bubbles) using the s parameter.
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 19]
sizes = [50, 200, 300, 100, 250]

plt.scatter(x, y, s=sizes, alpha=0.5, c='blue')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bubble Plot (using scatter)")
plt.show()
```
