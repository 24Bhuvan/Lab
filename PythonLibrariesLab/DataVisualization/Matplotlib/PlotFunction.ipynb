```python
#The plot() function is used to draw line graphs by connecting data points with straight lines.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('iris')
df.head()
```

```python
plt.plot(df['sepal_length'], df['sepal_width'], marker='o', linestyle='--',linewidth=2, markersize=6)
```

```python
species_list = df['species'].unique()

# Plot multiple lines
for species in species_list:
    subset = df[df['species'] == species]
    plt.plot(subset['sepal_length'], subset['sepal_width'], marker='o', label=species)
```
