```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("flights")
pivot_df = df.pivot_table(index="month", columns="year", values="passengers")

sns.heatmap(
    data=pivot_df.astype(int),   
    cmap="YlGnBu",
    annot=True,
    fmt="d",
    linewidths=0.5,
    linecolor="white",
    cbar=True
)

plt.show()
```
