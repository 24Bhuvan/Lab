```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.get_dataset_names()
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
sns.scatterplot(
    x='total_bill', 
    y='tip', 
    data=df, 
    hue='time', 
    style='day', 
    size='smoker',
    palette='deep'
)
```
