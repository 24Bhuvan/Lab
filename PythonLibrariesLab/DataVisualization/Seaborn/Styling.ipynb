```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.get_dataset_names()
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
sns.scatterplot(
    x='total_bill', 
    y='tip', 
    data=df, 
    hue='time', 
    style='day', 
    size='smoker',
    palette='deep'
)
#1
#sets the background style of Seaborn plots, controlling grid visibility, axes, and overall aesthetic. It affects all plots globally unless overridden.
sns.set_style(style='darkgrid')
# style: One of the predefined styles:
# 'white'  plain background, no grid
# 'whitegrid'  white background with grid (good for data plots)
# 'dark'  dark background, no grid
# 'darkgrid'  dark background with grid
# 'ticks'  minimal style with ticks on axes

#2
#removes one or more spines (the top and right axes lines) from a plot, creating a cleaner, minimalistic appearance.
sns.despine(top=True, right=True, left=False, bottom=False) 

#3
#sets the scaling of plot elements (like labels, lines, and markers) to make plots suitable for different contexts  e.g., presentation, paper, or notebook display.
sns.set_context(context='notebook')
# Must-Know Parameters:
# context: One of the predefined contexts:
# 'paper'  smaller elements, ideal for publications
# 'notebook'  default size for notebooks
# 'talk'  larger elements for presentations
# 'poster'  very large elements for big posters or slides
```
